```python
import os
import glob2 as glob
import shutil
```

```python
data_dir = r'D:\QXAS_May2022\MBP46B_Mo3Nb4O44_001_UP'# where is ALL the data stores?

output_dir = r'D:\QXAS_May2022\Subset_Data' #make me - Where can I save a copy of a 100 file subset of the data?
```

```python
files_temp = glob.glob(data_dir+'/**', recursive=True)
```

```python
files = []

for line in files_temp:
    if not os.path.isdir(line):
        files.append(line)
```

```python
print(len(files_temp))
print(len(files))
```

```python
len(files)//100
```

```python
files_to_open = []

for i in range(100):
    print(i*155)
    files_to_open.append(files[i*155])
```

```python
len(files_to_open)
```

```python
for line in files_to_open:
    base = os.path.basename(line)
    
    dest = os.path.join(output_dir, base)
    
    shutil.copyfile(line, dest)
```
